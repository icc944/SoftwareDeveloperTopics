# Algoritmos de Ordenamiento

Los algoritmos de ordenamiento nos permite, como su nombre lo dice, ordenar información de una manera especial basándonos en un criterio de ordenamiento.

En la computación el ordenamiento de datos cumple un rol muy importante, ya sea como un fin en sí o como parte de otros procedimientos más complejos. Se han desarrollado muchas técnicas en este ámbito, cada una con características específicas, y con ventajas y desventajas sobre las demás.

#### 1.- Algoritmo de ordenamiento por Bubble

El algoritmo de ordenamiento por el método de burbuja es un algoritmo simple y básico utilizado para ordenar elementos en una lista o arreglo. El concepto detrás de este algoritmo es comparar elementos adyacentes y, si están en el orden incorrecto, intercambiarlos hasta que toda la lista esté ordenada.

Ventajas del método de burbuja:

- **Implementación simple:** Es fácil de entender e implementar, lo que lo hace útil para fines educativos y conceptuales.
- **Requisitos de memoria:** Solo se necesita una cantidad constante de memoria adicional para realizar intercambios, lo que lo hace eficiente en términos de uso de memoria.
- **Puede detectar si la lista ya está ordenada:** Si la lista ya está en orden, el algoritmo puede detenerse temprano, lo que puede ser una ventaja en algunos casos.

Desventajas del método de burbuja:

- **Ineficiente en términos de tiempo:** El algoritmo tiene una complejidad de tiempo promedio y peor caso de O(n^2), lo que significa que puede volverse muy lento para listas grandes. Esto lo hace poco práctico para conjuntos de datos extensos.
- **No es óptimo:** Hay algoritmos de ordenamiento más eficientes, como QuickSort, MergeSort y HeapSort, que superan ampliamente al método de burbuja en términos de velocidad en la mayoría de los casos.
- **No es adecuado para listas grandes:** Debido a su ineficiencia, el método de burbuja no es una opción recomendada para ordenar grandes cantidades de datos.

Explicación: https://youtu.be/pqZ04TT15PQ?list=PLfBtpqIBIz7qyXl8TK8KPHYylRVlvIFY8

```python
def bubble_sort(arr):
  for i in range(len(arr)):
    for j in range(len(arr)-i-1):
      if arr[j] > arr[j+1]:
        arr[j], arr[j+1] = arr[j+1], arr[j]
  return arr
```

```python
nums = [4,5,3,6,5,1,3,8,7,9,5,4,2]
bubble_sort(nums)
```

#### 2.- Algoritmo de ordenamiento por Selección

El algoritmo de ordenamiento por selección es otro método de clasificación simple pero eficiente. Aquí tienes algunas ventajas y desventajas asociadas con este algoritmo:

Ventajas:

- **Simpleza de implementación:** Al igual que el algoritmo de inserción, el algoritmo de selección es fácil de entender y programar. Implica comparaciones y movimientos sencillos de elementos.

- **Ordenamiento en el lugar:** El ordenamiento por selección ordena los elementos en la misma lista sin requerir memoria adicional para copias temporales, lo que puede ser beneficioso en términos de uso de memoria.

- **Buen rendimiento en listas pequeñas:** Al igual que el algoritmo de inserción, el algoritmo de selección puede ser más rápido que algoritmos más complejos en listas pequeñas o casi ordenadas.

- **Cantidad fija de intercambios:** A diferencia de algunos otros algoritmos de ordenamiento, el algoritmo de selección realiza un número fijo de intercambios (n-1 intercambios en una lista de n elementos), lo que puede ser ventajoso en situaciones donde el número de escrituras es costoso.

Desventajas:

- Ineficiente para listas grandes:** El mayor inconveniente del algoritmo de selección es su ineficiencia en listas grandes. Su complejidad promedio y peor caso es cuadrática (O(n^2)), lo que lo hace inadecuado para tareas donde se manejan grandes volúmenes de datos.

- No se adapta a los datos:** El algoritmo de selección no se adapta a los patrones específicos de los datos. Realiza las mismas comparaciones y movimientos independientemente de las características de los datos, lo que lo hace menos eficiente en comparación con algoritmos adaptativos como el algoritmo de ordenamiento rápido.

- No es estable:** En su forma básica, el algoritmo de selección no es estable. Esto significa que puede cambiar el orden relativo de elementos iguales en la lista.

- No es adecuado para listas casi ordenadas:** Aunque el algoritmo de selección puede ser eficiente en listas pequeñas casi ordenadas, no es tan efectivo como otros algoritmos diseñados específicamente para aprovechar esta propiedad, como el algoritmo de inserción.

Explicación: https://youtu.be/Myy-eU-SWbE?list=PLfBtpqIBIz7qyXl8TK8KPHYylRVlvIFY8

```python
def selection_sort(arr):
  for i in range(0,len(arr)):
    idx_minimo=i
    for j in range(i+1, len(arr)):
      if arr[idx_minimo] > arr[j]: # En caso de que exista un valor más pequeño
        idx_minimo = j

    # Intercambiamos valores usando la asignacion en paralelo
    # Mover el valor minimo encontrado a la posicion actual y el actual la posicion del minimo
    arr[i], arr[idx_minimo] = arr[idx_minimo], arr[i]
  return arr
```

```python
#%%timeit
nums=[4,7,3,8,4,1]
selection_sort(nums)
```

#### 3.- Algoritmo de ordenamiento por inserción

El algoritmo de ordenamiento por inserción es un método simple y eficiente para ordenar elementos en una lista. A continuación, se presentan algunas ventajas y desventajas de este algoritmo:

Ventajas:

- **Simpleza de implementación:** El algoritmo de ordenamiento por inserción es bastante fácil de entender e implementar. Solo involucra comparaciones y movimientos simples de elementos en la lista.

- **Eficiente para listas pequeñas:** En listas pequeñas o casi ordenadas, el algoritmo de inserción puede ser más rápido que otros algoritmos de ordenamiento más complejos, como el algoritmo de ordenamiento rápido (quicksort) o el algoritmo de mezcla (merge sort).

- **Ordenamiento en el lugar:** El ordenamiento por inserción ordena los elementos en la misma lista, sin requerir memoria adicional para copias temporales, lo que puede ser útil en situaciones con restricciones de memoria.

- **Eficiente para datos parcialmente ordenados:** Cuando la mayoría de los elementos en la lista ya están en orden, el algoritmo de inserción puede ser muy eficiente, ya que solo necesita hacer algunas comparaciones y movimientos mínimos.

Desventajas:

- **Ineficiencia en listas grandes:** A medida que el tamaño de la lista aumenta, el algoritmo de ordenamiento por inserción se vuelve menos eficiente en comparación con otros algoritmos de ordenamiento más avanzados, ya que su complejidad promedio es cuadrática (O(n^2)).

- **No es óptimo para listas desordenadas:** En listas desordenadas o inversamente ordenadas, el algoritmo de inserción puede requerir un gran número de comparaciones y movimientos, lo que lo hace más lento en comparación con otros algoritmos de ordenamiento.

- **Falta de adaptabilidad:** El algoritmo de inserción no se adapta bien a la estructura de los datos. No puede aprovechar patrones específicos en los datos para mejorar su eficiencia, como lo hacen otros algoritmos como el algoritmo de ordenamiento rápido.

- **No es estable en su forma básica:** La versión más simple del algoritmo de inserción no es estable, lo que significa que puede cambiar el orden relativo de elementos iguales en la lista.

Explicación: https://youtu.be/6GU6AGEWYJY?si=0fSzfSZ2TOEZ3Wmq

```python
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]  # Elemento actual a ser insertado en su posición correcta
        j = i - 1

        # Mover los elementos mayores que key una posición hacia adelante
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1

        arr[j + 1] = key  # Insertar el elemento en su posición correcta
    return arr
```

```python
#%%timeit
arr = [4,7,3,8,4,1]
insertion_sort(arr)
```

#### 4.- Algoritmo de ordenamiento por Merge Sort

El algoritmo de ordenamiento Merge Sort es un método de ordenamiento eficiente que utiliza el enfoque de "dividir y conquistar" para ordenar una lista o un arreglo. Este algoritmo divide repetidamente la lista en mitades más pequeñas, ordena esas mitades y luego las combina en una lista ordenada final. Aquí te explico cómo funciona:

- **Dividir:** La lista no ordenada se divide en dos mitades aproximadamente iguales.
- **Conquistar:** Cada mitad se ordena recursivamente utilizando el mismo algoritmo de Merge Sort.
- **Combinar:** Las dos mitades ordenadas se combinan en una lista ordenada. Este proceso de combinación implica comparar los elementos de ambas mitades y colocarlos en orden en la lista final.

Ventajas del algoritmo Merge Sort:

- **Eficiencia en tiempo:** El Merge Sort tiene una complejidad de tiempo de O(n log n) en todos los casos, lo que lo hace eficiente incluso para conjuntos de datos grandes. Esto lo hace adecuado para aplicaciones en las que se requiere un ordenamiento estable y eficiente.
- **Rendimiento predecible:** Aunque Merge Sort puede no ser el algoritmo más rápido en términos de tiempo constante, su rendimiento consistente y predecible lo hace útil en situaciones donde la eficiencia es fundamental.
- **Estabilidad:** El Merge Sort es un algoritmo estable, lo que significa que mantiene el orden relativo de elementos iguales. Esto es importante en muchas aplicaciones.


Desventajas del algoritmo Merge Sort:

- **Requisitos de memoria:** El Merge Sort necesita más memoria adicional para almacenar las mitades descompuestas antes de combinarlas. En aplicaciones con limitaciones de memoria, esto puede ser una desventaja.
- **No siempre óptimo para listas pequeñas:** Aunque Merge Sort es muy eficiente para listas grandes, su sobrecarga en la división y combinación puede hacerlo un poco menos eficiente que otros algoritmos de ordenamiento (como QuickSort) para listas pequeñas.
- **Implementación más compleja:** Aunque el concepto de dividir y combinar es intuitivo, la implementación real del Merge Sort puede ser más complicada que la de algoritmos más simples como Bubble Sort.


Explicación: https://www.youtube.com/watch?v=ACFZn_xQcz8&list=PLfBtpqIBIz7qyXl8TK8KPHYylRVlvIFY8&index=4&pp=iAQB

```python
def do_merge(left_arr,right_arr):
  i=0
  j=0
  arr_sorted=[]

  # Recorrer ambos sin agotar valores de uno o de otro
  while i<len(left_arr) and  j<len(right_arr):
    if left_arr[i] < right_arr[j]:
      arr_sorted.append(left_arr[i])
      i+=1
    else:
      arr_sorted.append(right_arr[j])
      j+=1;

  while i<len(left_arr):
    arr_sorted.append(left_arr[i])
    i+=1

  while j<len(right_arr):
    arr_sorted.append(right_arr[j])
    j+=1

  return arr_sorted
```

```python
def merge_sort(arr):
  # Caso base
  if len(arr) == 1: return arr

  middle = len(arr)//2
  left_arr=arr[:middle]
  right_arr=arr[middle:]

  sorted_left_arr=merge_sort(left_arr)
  sorted_right_arr=merge_sort(right_arr)

  return do_merge(sorted_left_arr,sorted_right_arr)
```

```python
nums=[2,5,3,1,5,7,5,8,4,2,8,9,5,4,7,5]
merge_sort(nums)
```

#### 5.- Algoritmo de ordenamiento por Quick Sort

Quicksort es un algoritmo de ordenamiento eficiente en la mayoría de los casos y es ampliamente utilizado en la práctica. Sin embargo, se debe tener cuidado al elegir el pivote para evitar el peor caso y se debe considerar el contexto en el que se utiliza, especialmente cuando se necesita un orden estable.

**Ventajas**

- **Eficiencia promedio:** En general, Quicksort es uno de los algoritmos de ordenamiento más eficientes en términos de tiempo promedio. Su complejidad promedio es O(n log n), lo que lo hace adecuado para manejar grandes cantidades de datos de manera rápida.

- **Dividir y conquistar:** Quicksort utiliza la estrategia "dividir y conquistar", lo que significa que divide el problema en subproblemas más pequeños y los resuelve por separado. Esto reduce el tiempo de ejecución al trabajar en problemas más manejables.

- **Uso mínimo de memoria:** A diferencia de algoritmos como Merge Sort, Quicksort no requiere memoria adicional para almacenar sublistas. En cambio, opera directamente en la lista original, lo que reduce la sobrecarga de memoria.

- **Adaptable:** La elección del pivote puede afectar el rendimiento, pero es posible implementar estrategias para mejorar las elecciones del pivote, como seleccionar el pivote de manera aleatoria o medianas de tres.

- **Buen rendimiento en la práctica:** Aunque su complejidad en el peor caso puede ser O(n^2) si se elige un pivote desfavorable, en la práctica, Quicksort suele superar a otros algoritmos de ordenamiento como Bubble Sort, Insertion Sort y Selection Sort.

**Desventajas**
- **Peor caso desfavorable:** Si no se maneja adecuadamente la elección del pivote, Quicksort puede tener un rendimiento deficiente en su peor caso, especialmente si se recibe una lista que ya está ordenada o casi ordenada. Esto puede llevar a una complejidad O(n^2).

- **No estable:** Quicksort no garantiza el orden relativo de elementos iguales en la lista, lo que significa que no es un algoritmo estable. Esto puede ser problemático en algunas aplicaciones.

- **Complejidad de implementación:** Quicksort es más complicado de implementar en comparación con algoritmos más simples como Bubble Sort o Insertion Sort. La elección del pivote y el proceso de particionamiento pueden ser difíciles de entender y depurar.

- **Necesidad de aleatorización:** Para evitar el peor caso desfavorable, a menudo es recomendable introducir algún nivel de aleatorización en la elección del pivote. Esto puede añadir complejidad al código.

Explicació: https://www.youtube.com/watch?v=UrPJLhKF1jY&list=PLfBtpqIBIz7qyXl8TK8KPHYylRVlvIFY8&index=5&pp=iAQB

Implementacion sin uso de apuntadores

```python
def quick_sort(arr):
  if len(arr) <=1:
    return arr

  pivot = arr[len(arr)//3]
  left = [x for x in arr if x < pivot]
  middle = [x for x in arr if x == pivot]
  right = [x for x in arr if x > pivot]

  return quick_sort(left) + middle + quick_sort(right)
```

```python
arr = [3,4,5,8,4,2,7,5,1,9]
quick_sort(arr)
```

Implementacion con apuntadores

```python
def quick_sort(arr):
    def _quick_sort(arr, i_min, i_max):
        if i_min < i_max:
            i_pivot = partition(arr, i_min, i_max)
            _quick_sort(arr, i_min, i_pivot - 1)
            _quick_sort(arr, i_pivot + 1, i_max)


    def partition(arr, i_min, i_max):
        pivot = arr[i_min]
        i = i_min + 1
        j = i_max

        while True:
            while i <= j and arr[i] <= pivot:
                i += 1
            while arr[j] >= pivot and j >= i:
                j -= 1
            if j < i:
                break
            else:
                arr[i], arr[j] = arr[j], arr[i]

        arr[i_min], arr[j] = arr[j], arr[i_min]
        return j

    _quick_sort(arr, 0, len(arr) - 1)
```

```python
arr = [8,7,9,3,4,5,8,2,7,5,9]
quick_sort(arr)
```

#### 6.- Algoritmo de ordenamiento por Heap Sort

Heap Sort es un algoritmo de ordenación basado en la estructura de datos llamada "montículo" o "heap". Es un algoritmo de ordenación por comparación y es muy eficiente en términos de tiempo de ejecución, ya que tiene una complejidad de tiempo de O(n log n) en el peor de los casos.

Aquí está una explicación paso a paso del algoritmo Heap Sort:

**Construcción del montículo (Heapify):**

- El primer paso es convertir la matriz no ordenada en un montículo (heap) en su forma original. Un montículo es una estructura de datos en la que el elemento en cada nodo es mayor (en un montículo máximo) o menor (en un montículo mínimo) que sus hijos.
- Se inicia la construcción del montículo desde el último nodo que tiene hijos (n // 2 - 1) hasta la raíz del montículo.
- En cada paso, se compara el nodo actual con sus hijos y se intercambia si es necesario para que el montículo cumpla con la propiedad del montículo.
- Después de la construcción del montículo, el elemento más grande (en un montículo máximo) estará en la raíz del montículo.


**Ordenación:**

- Una vez que se ha construido el montículo, el elemento más grande está en la raíz del montículo (posición 0 de la matriz).
- Se intercambia este elemento con el último elemento no ordenado en la matriz.
- Luego, se reduce el tamaño del montículo (ignorando el último elemento) y se restaura la propiedad del montículo.
- Repetir el paso anterior hasta que todos los elementos estén ordenados.
- El resultado será una matriz ordenada en orden ascendente (o descendente, dependiendo de si se utiliza un montículo máximo o mínimo).

El algoritmo Heap Sort tiene un buen rendimiento en términos de tiempo y es estable en su complejidad de tiempo O(n log n) en el peor de los casos, lo que lo hace útil para grandes conjuntos de datos. Sin embargo, en la práctica, puede ser menos eficiente que algunos otros algoritmos de ordenación como QuickSort o Merge Sort para conjuntos de datos pequeños debido a la constante de tiempo más alta asociada con las operaciones de montículo.

```python
def heapify(arr, n, i):
    # Inicializar el índice más grande como la raíz del montículo
    largest = i
    left = 2 * i + 1
    right = 2 * i + 2

    # Comprobar si el hijo izquierdo existe y es mayor que la raíz
    if left < n and arr[left] > arr[largest]:
        largest = left

    # Comprobar si el hijo derecho existe y es mayor que la raíz
    if right < n and arr[right] > arr[largest]:
        largest = right

    # Cambiar la raíz si es necesario
    if largest != i:
        arr[i], arr[largest] = arr[largest], arr[i]  # Intercambiar
        # Llamar recursivamente a heapify en el subárbol afectado
        heapify(arr, n, largest)

def heapSort(arr):
    n = len(arr)

    # Construir un montículo máximo (heap)
    for i in range(n // 2 - 1, -1, -1):
        heapify(arr, n, i)

    # Extraer elementos uno por uno desde el montículo
    for i in range(n - 1, 0, -1):
        arr[i], arr[0] = arr[0], arr[i]  # Intercambiar el elemento máximo con el último
        heapify(arr, i, 0)

# Ejemplo de uso
arr = [12, 11, 13, 5, 6, 7]
heapSort(arr)
print("Arreglo ordenado:")
print(arr)
```

#### 1.- Algoritmo de ordenamiento por Conteo

El algoritmo de ordenamiento por conteo es un método que se utiliza para ordenar una lista de elementos en función de la cantidad de veces que aparece cada elemento en la lista. Aunque este algoritmo tiene algunas ventajas, también tiene ciertas limitaciones. Aquí tienes una lista de ventajas y desventajas del ordenamiento por conteo:

**Ventajas:**

- **Eficiente para rangos pequeños de valores:** El ordenamiento por conteo es extremadamente eficiente cuando se trabaja con un rango pequeño de valores posibles. Si los valores a ordenar están limitados a un rango pequeño, el algoritmo puede ser mucho más rápido que otros métodos de ordenamiento como el quicksort o el mergesort.

- **No depende de comparaciones:** A diferencia de otros algoritmos de ordenamiento que se basan en comparaciones entre elementos, el ordenamiento por conteo no necesita realizar comparaciones. Esto significa que su rendimiento no se ve afectado por la naturaleza de los datos (ordenados, parcialmente ordenados o desordenados).

- **Estable:** El ordenamiento por conteo es un algoritmo estable, lo que significa que si hay elementos repetidos con el mismo valor, su orden relativo se mantendrá después de la ordenación.

**Desventajas:**

- **Ineficiente para rangos grandes:** A medida que el rango de valores posibles aumenta, el rendimiento del ordenamiento por conteo disminuye significativamente. Requiere una cantidad considerable de memoria y tiempo para crear y manipular un arreglo de conteo que abarque el rango completo.

- **Requiere valores enteros no negativos:** El ordenamiento por conteo solo funciona con valores enteros no negativos. No es adecuado para ordenar elementos con valores en coma flotante o valores negativos sin modificaciones adicionales.

- **Espacio de memoria:** El algoritmo requiere una cantidad significativa de espacio de memoria adicional para almacenar el arreglo de conteo. Esto puede ser una desventaja en sistemas con limitaciones de memoria.

- **No es adaptable:** A diferencia de otros algoritmos como el quicksort que pueden ser adaptativos y optimizar su rendimiento según la disposición de los datos, el ordenamiento por conteo no se adapta a la distribución de los valores.



Explicación: https://youtu.be/x-Yo8zAy9Uw?list=PLfBtpqIBIz7qyXl8TK8KPHYylRVlvIFY8

```python
def count_sort(arr):
  output = [0] * len(arr)
  count = [0] * (max(arr)+1)

  # Conteo de valores
  for num in arr:
    count[num] += 1

  # Hacer suma acumulada de los conteos
  for i in range(1, max(arr)+1):
    count[i] += count[i-1]

  # Reubicaciones de valores
  for i in range(0,len(arr)):
    valor2sort = arr[i]
    pos2insert = count[valor2sort] - 1
    output[pos2insert] = valor2sort
    count[valor2sort] -= 1

  return output

count_sort([1,5,7,4,4,3,1])
```

#### 2.- Algoritmo de ordenamiento por Radix Sort

El algoritmo Radix Sort es un algoritmo de ordenación que se basa en el ordenamiento de elementos a través de sus dígitos individuales. En lugar de comparar elementos directamente, como en otros algoritmos de ordenación como el Quicksort o el Merge Sort, Radix Sort trabaja con los dígitos en posiciones específicas (desde el dígito menos significativo al más significativo) para organizar los elementos. Funciona especialmente bien cuando se ordenan números enteros, cadenas de caracteres o cualquier otro tipo de dato que pueda ser descompuesto en componentes individuales.



**Ventajas**

- **Estabilidad:** Radix Sort es estable, lo que significa que mantiene el orden relativo de los elementos con valores iguales. Esto puede ser importante en ciertos casos en los que se necesita preservar la ordenación original.

- **Eficiencia en ciertos casos:** En ciertas circunstancias, Radix Sort puede ser más rápido que otros algoritmos de ordenación, especialmente cuando se trabaja con un gran número de elementos y un rango relativamente pequeño de valores.

- **No depende de comparaciones:** A diferencia de algoritmos como Quicksort o Merge Sort, Radix Sort no depende de comparaciones directas entre elementos, lo que puede hacer que sea más eficiente en ciertos casos.



**Desventajas del algoritmo Radix Sort:**

- **Complejidad en valores grandes o negativos:** Radix Sort puede volverse menos eficiente cuando se trata con valores grandes o negativos, ya que necesita iterar sobre múltiples dígitos y realizar múltiples pasos de ordenación.

- **Uso de memoria:** En implementaciones tradicionales, Radix Sort puede requerir una cantidad significativa de memoria adicional para almacenar las listas auxiliares de ordenación.

- **No siempre la mejor opción:** Aunque Radix Sort puede ser eficiente en ciertos casos, no es la mejor opción en todas las situaciones. Algoritmos como Quicksort y Merge Sort pueden ser más rápidos en conjuntos de datos más pequeños o en ciertos patrones de datos.

Explicación: https://www.youtube.com/watch?v=W_euZjKoHkM&t=28s

```python
# Radix sort in Python
def countingSort(array, place):
    size = len(array)
    output = [0] * size
    count = [0] * 10

    # Calculate count of elements
    for i in range(0, size):
        index = array[i] // place
        count[index % 10] += 1

    # Calculate cumulative count
    for i in range(1, 10):
        count[i] += count[i - 1]

    # Place the elements in sorted order
    i = size - 1
    while i >= 0:
        index = array[i] // place
        output[count[index % 10] - 1] = array[i]
        count[index % 10] -= 1
        i -= 1

    for i in range(0, size):
        array[i] = output[i]
```

```python
def radix_sort(array):
    # Get maximum element
    max_element = max(array)

    # Apply counting sort to sort elements based on place value.
    place = 1
    while max_element // place > 0:
        countingSort(array, place)
        place *= 10
```

```python
data = [121, 232, 564, 23, 1,432, 45, 788]
radix_sort(data)
print(data)
```
